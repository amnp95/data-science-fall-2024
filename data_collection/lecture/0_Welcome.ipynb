# Welcome!

Welcome to the Erdős Institute's data science educational content!

In this short notebook we will:

* Look at the organization and format of our materials
* Understand which materials are part of the "core" content and which are more optional.
* Discuss the prerequisites needed to make the most out of this experience.

### Structure of the Repository

The repository is organized as follows

```
.
└── data/
└── week-n-descriptive-name/
    ├── lecture-n/
    ├── problem-session-n/
    ├── math-n/
    ├── optional-extra-practice-n/
    └── optional-additional-content-n/
```

### Core Content:  Lecture Notebooks

Each topic we cover will have one or more associated lecture notebooks. These are the notebooks which will be covered in the Live Lecture each week. In lecture notebooks you will find a lecture-style presentation of topics. In particular, there will be writing that explains techniques and provide context as well as python code that demonstrates how to implement said techniques. Each notebook has both a "blank" and "complete" version.  The blank version omits some of the code so that you can practice writing code during the lecture. If you are consuming this content asynchronously, you can find pre-recorded lecture videos on the Erdős Institute website.

### Core Content: Problem Session Notebooks

Problem session notebooks give you the opportunity to practice the material we learn in lecture. You should attempt these during the problem session.  Try not to look at the solutions until after you have completed the "blank" versions yourself!

Most problem sessions also have "optional prep notebooks".  As their name implies:  these are optional!  They are short refreshers on the python skills needed to complete the problem session. 

### Optional Content:  Math Hour Notebooks

Math hour notebooks cover the mathematics underlying some of the data science techniques we learn in the bootcamp.  You do not need to understand the inner workings of these algorithms in this depth to be a great data scientist!  

### Optional Content:  Optional Extra Practice

You can think of this as extra problem sessions.  You can use these problems to sharpen your skills!  If you form a study group you might also enjoy giving yourself an extra weekly problem session where you cover some of these notebooks.

### Optional Content:  Optional Additional Content

These are lecture notebooks which cover content which is relevant to the topic of the week, but which we will not have time to cover in live lecture.

## Prerequisites

### Coding

In these notebooks we assume that you know the basics of Python. In particular you should be familiar with:
- `pandas`,
- `numpy` and
- `matplotlib`.
If you are not familiar with python work through the Erd&#337;s Institute's python prep materials, available on our website.

### Math and Statistics

For those of you looking to understand the math presented in these notebooks an absolute minimum includes:
- Multivariable Calculus (primarily partial derivatives and the Jacobian matrix),
- Some linear algebra,
- Basic probability theory and
- Basic statistics.

If you want a very brief refresher on these check out the First Steps section of the website for review slides.

If you want a more in depth look at linear algebra and multivariable calculus, you can work through the content in the week_00_math_prereq_crash_course folder.
